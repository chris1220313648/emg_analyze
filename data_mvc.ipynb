```python
filePath = './data/'
savePath = './data_mvc/'
count = 0
featureArr = []
mvc = pd.read_csv('./max/max.csv', header=1).astype(float)
mvc=np.array(mvc).T
# max_0=np.max(mvc[2])
# min_0=np.min(mvc[2])
# max_1=np.max(mvc[3])
# min_1=np.min(mvc[3])
# max_2=np.max(mvc[4])
# min_2=np.min(mvc[4])
# max_3=np.max(mvc[5])
# min_3=np.min(mvc[5])
# max_4=np.max(mvc[6])
# min_4=np.min(mvc[6])
# max_5=np.max(mvc[7])
# min_5=np.min(mvc[7])
for root, dirs, files in os.walk(filePath):
    for file in files:
        path = os.path.join(root, file)
        print(path)
        sPath = savePath + path.split("/")[-2] + "/" +  file
        data = pd.read_csv(path, header=1).astype(float)
        # print(data.head())
        data = np.array(data).T
        # print(type(data[5]))
        # print(data.shape)
        
        for i in range(2, 8):
            min_val = np.min(mvc[i])
            # print(min_val)
            max_val = np.max(mvc[i])
            # print(max_val)
            range_val = max_val - min_val  # -
            # print(range_val)
            if range_val != 0:  # 0
                data[i] = (data[i] - min_val) / range_val
                print(data[i])
            else:
                # range_val00
                data[i] = 0  # 
            # print(data[i])
        pd.DataFrame(data).T.to_csv(sPath, header=None, index=None)
```

```python
import pandas as pd
import numpy as np
# 
mvc = pd.read_csv('./max/max.csv', header=1).astype(float)
mvc=np.array(mvc).T
# 
max_values = []
min_values = []

# 
for i in range(2,8):
    max_values.append(np.max(mvc[i]))
    min_values.append(np.min(mvc[i]))

#  DataFrame 
max_min_df = pd.DataFrame({
    'column': [i for i in range(6)],
    'max_value': max_values,
    'min_value': min_values
})

# 
save_path = './max/min_max_values.csv'

#  CSV 
max_min_df.to_csv(save_path, index=False)

print(f"Max and min values saved to {save_path}")
```
